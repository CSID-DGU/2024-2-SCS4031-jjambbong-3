```python
path = '/content/drive/MyDrive/tamtam/국내 지역별 관광명소 데이터.csv'
place = pd.read_csv(path, encoding='utf-8')
```

```python
data = place.copy()
```

```python
data.head()
```

```python
## 컬럼 정의 홈페이지 참고
data.info()
```

```python
# '시도명' 컬럼에서 '제주'를 포함하는 행만 필터링
jeju_data = data[data['CTPRVN_NM'].str.contains("제주")]
```

```python
# 필요한 컬럼만 선택 (예: ID, POI_NM, CTPRVN_NM, SIGNGU_NM, CL_NM, LC_LO, LC_LA)
selected_columns = ['POI_NM', 'CTPRVN_NM', 'SIGNGU_NM','RDNMADR_NM','LEGALDONG_NM','LNBR_NO', 'CL_NM', 'LC_LO', 'LC_LA']
jeju_data_filtered = jeju_data[selected_columns].drop_duplicates()

# 필터링된 데이터 확인
jeju_data_filtered.head()
```

```python
jeju_data_filtered['CL_NM'].unique()
```

```python
jeju_data_filtered = jeju_data_filtered[jeju_data_filtered['CL_NM'] != '항공사/여행사']
jeju_data_filtered = jeju_data_filtered[jeju_data_filtered['CL_NM'] != 'N']
```
