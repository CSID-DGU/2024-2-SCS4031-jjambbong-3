```python
import os
from pinecone import Pinecone, ServerlessSpec
from sentence_transformers import SentenceTransformer

# Pinecone 초기화
pc = Pinecone(
    api_key=PINECONE_API_KEY
)

# 인덱스 생성
if 'tamtam2' not in pc.list_indexes().names():
    pc.create_index(
        name='tamtam2',
        dimension=384,  # 384차원으로 설정
        metric='cosine',  # 코사인 유사도를 기준으로 검색
        spec=ServerlessSpec(
            cloud='aws',
            region='us-east-1'  # 원하는 리전 설정
        )
    )
```

```python
# SentenceTransformer 모델 로드
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
# 기존 인덱스 참조
index = "tamtam2"
```

```python
# 기존 인덱스 참조
index = pc.Index(index)  # 기존 인덱스를 참조
print(f"Index '{index}' is ready for use.")
```

```python
import re

# 1. 결측값 처리
final_data = final_data.fillna({
    "review": "리뷰 없음",  # "리뷰 없음"으로 채워 사용자 이해도를 높임
    "rating": 0.0,  # 평점은 0.0으로 대체
    "latitude": 0.0,  # 위도와 경도는 (0.0, 0.0)으로 설정
    "longitude": 0.0,
    "address": "주소 정보 없음",  # 주소가 없으면 "주소 정보 없음"으로 처리
    "name": "알 수 없음",  # 이름이 없으면 "알 수 없음"으로 설정
    "category": "기타",
    "theme": "미정"
})

# 2. 데이터 업로드
for _, row in final_data.iterrows():
    # 고유 ID 생성
    unique_id = f"{row['name']}_{row['category']}_{row['address']}".replace(" ", "_")
    unique_id = re.sub(r'[^\x00-\x7F]+', '', unique_id)  # 비 ASCII 문자 제거

    # Pinecone에 ID 존재 여부 확인
    response = index.fetch([unique_id])
    if unique_id in response['vectors']:  # ID가 벡터 목록에 존재하는지 확인
        continue  # 존재하면 다음 루프로 넘어감

    # 2.2 카테고리에 따른 쿼리 생성
    queries = []
    if row["category"] == "음식점":
        queries.extend([
            f"{row['name']}은 제주도 {row['theme']} 음식점 입니다.",
            f"제주도에서 {row['theme']}를 먹고 싶다면 {row['name']}을 추천합니다.",
            f"{row['address']} 근처에서 가볼만한 음식점은 {row['name']}입니다.",
            f"평점 {row['rating']}점으로 추천되는 음식점 {row['name']}."
        ])
    elif row["category"] == "카페":
        queries.extend([
            f"{row['name']}은 제주도 카페입니다.",
            f"제주도에서 맛있는 커피를 마시고 싶다면 {row['name']}을 추천합니다.",
            f"{row['address']} 근처에서 방문할 만한 카페는 {row['name']}입니다.",
            f"평점 {row['rating']}점으로 추천되는 카페 {row['name']}.",
            f"{row['name']}은 {row['review']} 라는 평을 받은 카페입니다."
        ])
    elif row["category"] == "관광지":
        queries.extend([
            f"{row['name']}은 {row['theme']} 테마의 제주도 관광지입니다.",
            f"제주도에서 가볼만한 관광지로 {row['name']}을 추천합니다.",
            f"{row['address']} 근처에서 방문할 수 있는 명소는 {row['name']}입니다."
        ])
    else:
        queries.append(f"{row['name']}")

    # 테마 관련 추가 쿼리
    if row["theme"] != "미정":
        queries.append(f"{row['name']}은 {row['theme']} 테마와 관련된 장소입니다.")
        queries.append(f"제주도에서 {row['theme']} 테마를 원한다면 {row['name']}을 고려해보세요.")

    # 2.3 텍스트 필드 생성 (쿼리 포함)
    text = f"{row['name']} - {row['category']}:\n" + "\n".join(queries)

    # 2.4 임베딩 생성
    embedding = model.encode(text).tolist()

    # 2.5 메타데이터 구성
    metadata = {
        "name": row["name"],
        "address": row["address"],
        "rating": row["rating"],
        "latitude": row["latitude"],
        "longitude": row["longitude"],
        "category": row["category"],
        "review": row["review"],
        "theme": row["theme"]
    }

    # 2.6 Pinecone 업로드
    index.upsert([{
        "id": unique_id,
        "values": embedding,
        "metadata": metadata
    }])

print("Data uploaded successfully!")
```

```python
query_text = "고기국수 맛집 추천해줘"
query_vector = model.encode(query_text).tolist()

# 음식점 카테고리로 필터링
results = index.query(
    vector=query_vector,
    top_k=5,              # 상위 5개 결과 반환
    include_metadata=True,
    filter={
        "category": "음식점"  # 음식점만 검색
    }
)

# 결과 출력
for match in results["matches"]:
    print("Score:", match["score"])
    print("Metadata:", match["metadata"])
```
