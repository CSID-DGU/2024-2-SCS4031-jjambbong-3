```python
data = jeju_attractions.copy()
```

```python
# 변수명 변경
data.columns = ['관광지명', '시/도명', '시/군/구명', '도로명 주소', '법정동명', '지번 번호', '분류명', '경도', '위도']

# 변경된 데이터프레임 확인
print(data)
```

```python
# 주소 결합
data['주소'] = data.apply(
    lambda row: f"{row['시/도명']} {row['시/군/구명']} {row['도로명 주소'] if pd.notna(row['도로명 주소']) else ''} {row['법정동명']} {row['지번 번호']}".strip(),
    axis=1
)

# 기존 열 삭제 (선택 사항)
data.drop(['시/도명', '시/군/구명', '도로명 주소', '법정동명', '지번 번호'], axis=1, inplace=True)
```

```python
data['카테고리'] = "관광지"
```

```python
# 테마 사전 정의
themes = {
    "자연": [
        "유명사적/유적지", "관광농원/허브마을", "천연기념물", "폭포/계곡",
        "해수욕장", "휴양림/수목원", "식물원", "동물원"
    ],
    "액티비티": [
        "레저/체험", "팜스테이", "야영장", "캠핑장",
        "캠핑홀리데이(캠핑)", "테마공원/대형놀이공원", "일반유원지/일반놀이공원", "아쿠아리움/대형수족관"
    ],
    "쇼핑": [
        "시장", "특산품", "먹거리/패션거리", "지역축제", "정보화마을"
    ],
    "여유로운 힐링": [
        "국립공원", "온천지역", "온천", "휴양림/수목원"
    ],
    "문화/예술/역사": [
        "박물관/전시관", "유명사적/유적지", "고택/생가/민속마을",
        "성/성터", "보물", "비/탑/문/각",
        "서원/향교/서당", "왕릉/고분", "드라마/영화촬영지",
        "관광안내소/매표소", "일반관광지", "영어마을"
    ]
}

# "분류명"에 따라 테마 매핑 함수 정의
def map_theme(category, themes):
    for theme, categories in themes.items():
        if category in categories:
            return theme
    return "기타"  # 매핑되지 않는 분류명에 대해 "기타" 반환

# 새로운 열 '테마' 추가
data['테마'] = data['분류명'].apply(lambda x: map_theme(x, themes))
```

```python
data = data.drop(columns=['분류명'])  # '분류명' 열 삭제
```
